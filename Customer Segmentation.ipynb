```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url="https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/customer_segmentation.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.drop('Address',axis=1, inplace=True)
df.head()
```

```python
X = df.values[:,1:]
X = np.nan_to_num(X)
norm_data = StandardScaler().fit_transform(X)
norm_data
```

```python
k = 3
k_means = KMeans(n_clusters = k)
k_means.fit(norm_data)
print(k_means.labels_)
```

```python
df['cluster'] = k_means.labels_
df.head()
```

```python
df.groupby('cluster').mean()
```

```python
area = np.pi * ( X[:, 1])**2
plt.scatter(X[:, 0], X[:, 3], s=area, c=k_means.labels_.astype(float), alpha=0.5)
plt.xlabel('Age', fontsize=18)
plt.ylabel('Income', fontsize=16)

plt.show()
```

```python
df.to_csv('Customer Segments') #Customer list with their segments
```
